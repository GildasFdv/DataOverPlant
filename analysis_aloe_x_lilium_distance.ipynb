```python

import pandas as pd
import plotly.graph_objects as go
```

```python
df_lilium = pd.read_csv('./data/21_01_2025_lilium.csv')
df_aloevera = pd.read_csv('./data/21_01_2025_aloevera.csv')
df = pd.concat([df_lilium, df_aloevera])
```

```python
def get_color(dist, plant):
    if plant == 'lilium':
        if dist == 5:
            return '#aed3e3'
        elif dist == 10:
            return '#009dcf'
        elif dist == 15:
            return '#007aa2'
        elif dist == 20:
            return '#004b57'
    elif plant == 'aloevera':
        if dist == 5:
            return '#fea1b6'
        elif dist == 10:
            return '#ff577b'
        elif dist == 15:
            return '#ff073d'
        elif dist == 20:
            return '#b20027'
```

```python

# Exemple de sous-ensembles de donnes
df_aloevera = df[df['plant'] == 'aloevera']
df_lilium = df[df['plant'] == 'lilium']

# Cration d'une figure Plotly
fig = go.Figure()

# Ajout de courbes pour chaque catgorie
for i in range(1, 5):
    df_aloevera_dist = df_aloevera[df_aloevera['distance'] == i * 5]
    df_lilium_dist = df_lilium[df_lilium['distance'] == i * 5]
    fig.add_trace(go.Scatter(
        x=df_aloevera_dist['frequency'], 
        y=df_aloevera_dist['gain'], 
        mode='lines',
        name=f'Aloevera {i * 5}cm',
        line=dict(color=get_color(i * 5, 'aloevera'))
    ))
    fig.add_trace(go.Scatter(
        x=df_lilium_dist['frequency'], 
        y=df_lilium_dist['gain'], 
        mode='lines',
        name=f'Lilium {i * 5}cm',
        line=dict(color=get_color(i * 5, 'lilium'))
    ))

# Mise en page personnalise
fig.update_layout(
    title="Frequence vs Puissance",
    xaxis_title="Frequence (Hz)",
    yaxis_title="Puissance (dB)",
    legend_title="Plant Type",
    template="plotly"
)

# Affichage interactif
fig.show()
```
