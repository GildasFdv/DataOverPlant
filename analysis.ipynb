```python
import pandas as pd
```

```python
df = pd.read_csv('./data/17_01_2025.csv')
```

```python
import plotly.graph_objects as go

# Exemple de sous-ensembles de données
df_cactus = df[df['plant'] == 'cactus']
df_aloevera = df[df['plant'] == 'aloevera']
df_lilium = df[df['plant'] == 'lilium']
df_succulente = df[df['plant'] == 'succulente']

# Création d'une figure Plotly
fig = go.Figure()

# Ajout de courbes pour chaque catégorie
fig.add_trace(go.Scatter(
    x=df_cactus['frequency'], 
    y=df_cactus['gain'], 
    mode='lines',
    name='Cactus'
))
fig.add_trace(go.Scatter(
    x=df_aloevera['frequency'], 
    y=df_aloevera['gain'], 
    mode='lines',
    name='Aloevera'
))
fig.add_trace(go.Scatter(
    x=df_lilium['frequency'], 
    y=df_lilium['gain'], 
    mode='lines',
    name='Lilium'
))
fig.add_trace(go.Scatter(
    x=df_succulente['frequency'], 
    y=df_succulente['gain'], 
    mode='lines',
    name='Succulente'
))

# Mise en page personnalisée
fig.update_layout(
    title="Frequency vs Gain",
    xaxis_title="Frequency (Hz)",
    yaxis_title="Gain (dB)",
    legend_title="Plant Type",
    template="plotly"
)

# Affichage interactif
fig.show()
```
